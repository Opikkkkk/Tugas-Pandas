```python
import pandas as pd

print(" DATA TOTAL SAMPAH KOTA/KABUPATEN DI JAWA BARAT")
df_sampah = pd.read_csv('data_sampah.csv')
df_sampah

```

```python
#Menghitung Jumlah Produksi Berdasarkan Tahun
tahun = 2015
df_tahun = df_sampah[df_sampah['tahun_penulisan'] == tahun]
jumlah_produksi = df_tahun['produksi_sampah_ton_per_hari'].sum()

print("TOTAL DATA SAMPAH PADA TAHUN 2015")
print(f"Jumlah Produksi Sampah pada Tahun 2015 Sebanyak {jumlah_produksi:.2f} Ton per Hari")

#DataFrame Baru
total_data = pd.DataFrame({
    'tahun_penulisan': [tahun],
    'total_produksi_sampah_ton_per_hari': [jumlah_produksi]
})
total_data.to_csv('jumlah_produksi_tahun_2015.csv', index=False)
total_data.to_excel('jumlah_produksi_tahun_2015.xlsx', index=False, sheet_name='Data Produksi Sampah Tahun 2015')

#Menghitung Jumlah Produksi Berdasarkan Tahun ke Tahun
print("\nDATA SAMPAH PER TAHUN")
total_per_tahun = {}


for index, row in df_sampah.iterrows():
    tahun = row['tahun_penulisan']
    produksi = row['produksi_sampah_ton_per_hari']
    
    if tahun in total_per_tahun:
        total_per_tahun[tahun] += produksi
    else:
        total_per_tahun[tahun] = produksi

data_output = []
for tahun, total in total_per_tahun.items():
    print(f"Total produksi sampah di Jawa Barat pada tahun {tahun} adalah {total:.2f} ton per hari.")
    data_output.append({'Tahun': tahun, 'Total Produksi (ton per hari)': total})

df_output = pd.DataFrame(data_output)
df_output.to_csv("total_sampah_per_tahun.csv", index=False)
df_output.to_excel("total_sampah_per_tahun.xlsx", index=False)


# Menjumlahkan Jumlah Produksi Sampah Berdasarkan Kota/Kabupaten Diikuti dengan Tahunnya
hasil = df_sampah.groupby(['nama_kabupaten/kota', 'tahun_penulisan'])['produksi_sampah_ton_per_hari'].sum().reset_index()
print("\nDATA BERDASARKAN KABUPATEN/KOTA DI SETIAP TAHUNNYA")

for index, row in hasil.iterrows():
    print(f"Pada {row['nama_kabupaten/kota']} di Tahun {row['tahun_penulisan']}, Total Produksi Sampah Adalah {row['produksi_sampah_ton_per_hari']:.2f} Ton per Hari.")

hasil.to_csv('jumlah_produksi_sampah_berdasarkan_kota.csv', index=False)
hasil.to_excel('jumlah_produksi_sampah_berdasarkan_kota.xlsx', index=False, sheet_name='Data_Produksi_Sampah_Berdasarkan_Kabupaten_Kota_Per_Tahun')
```
